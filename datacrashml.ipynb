```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Model(nn.Module):
  def __init__(self, in_features=7, h1 = 8, h2 = 9, out_features=55):
    super().__init__()
    self.fcl = nn.Linear(in_features, h1)
    self.fc2 = nn.Linear(h1, h2)
    self.out = nn.Linear(h2, out_features)

  def foward(self, x):
    #rectafied linear unit
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.out(x)

    return x
```
